```python
## handling holidays in the date time series dataset

## import these modules
from pandas.tseries.holiday import USFederalHolidayCalendar  ## this is just an example, not updated
from pandas.tseries.offsets import CustomBusinessDay

## instantiate a calendar
usb = CustomBusinessDay(calendar = USFederalHolidayCalendar())

# this will remove 4th july(holiday) + weekends from the result
pd.date_range(start="7/1/2017", end="7/21/2017", freq=usb)
```

```python
## creating custom holiday calendars
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday

class IndianHolidayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday("Diwali", month=10, day=21),
        Holiday("Dusherra", month=10, day=11)
    ]

myCalendar = CustomBusinessDay(calendar=IndianHolidayCalendar())
myCalendar
```

```python
pd.date_range(start="10/5/2017", end="10/25/2017", freq=myCalendar) # notice holidays are missing with business days
```

```python
## another approach to add holiday and custom days as week days, then use weekmask and holidays parameters
myCalendar = CustomBusinessDay(weekmask="Sun Mon Tue Wed Thu", holidays=["2017-07-04"])
pd.date_range(start="7/1/2017", end="7/25/2017", freq=myCalendar)
```
